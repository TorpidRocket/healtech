```python
# chr(32) -> chr(123)
ls = []
for i in range(32, 123):
    ls.append(chr(i))
print(ls)
len(ls)
```

```python
wrong_char = {'"', "'", ':', ';', '<', '=', '>', '?', '[', '\\', ']', '^', '_', '`', '(', ')', '+', '/',}
```

```python
n = len(ls)
for i in range(n-1, -1, -1):
    if ls[i] in wrong_char:
        del ls[i]
print(ls)
len(ls)
```
