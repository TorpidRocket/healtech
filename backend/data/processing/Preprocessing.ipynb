```python
import pandas as pd

df = pd.read_csv("doctor.csv")
print(df.head)
```

#Preprocessing

Checking for NaN

```python
nan_count = df['DP Score'].isna().sum()
print(nan_count)
```

```python
nan_count = df['NPV Value'].isna().sum()
print(nan_count)
```

```python
nan_count = df['Consult Fee'].isna().sum()
print(nan_count)
```

```python
nan_count = df['Years of Experience'].isna().sum()
print(nan_count)
```

```python
nan_count = df['Degree'].isna().sum()
print(nan_count)
```

```python
nan_count = df['Name'].isna().sum()
print(nan_count)
```

```python
nan_count = df['City'].isna().sum()
print(nan_count)
```

```python
nan_count = df['Speciality'].isna().sum()
print(nan_count)
```

```python
df = df.dropna(subset=['Name', 'City', 'Speciality', 'Degree', 'Years of Experience', 'Consult Fee'])
```

```python
print(df.count())
```

Cleaning Dataset

```python
cl_df = df.copy()
cl_df.head()
```

Dropping Location column

```python
cl_df = cl_df.drop(columns=['Location'])
cl_df.head()
```

```python
cl_df.count()
```

```python
cl_df = cl_df.dropna()
cl_df.count()
```

```python
cl_df.head()
```

```python
cl_df['DP Score'] = cl_df['DP Score'].str.rstrip('%').astype(float) / 20
cl_df.head()
```

```python
cl_df['Consult Fee'] = cl_df['Consult Fee'].str.replace('₹', '', regex=True).astype(float)
```

```python
cl_df.head()
```

```python
print(cl_df['Years of Experience'].apply(type).value_counts())
```

```python
cl_df = cl_df[cl_df['Years of Experience'].apply(lambda x: str(x).replace('.', '', 1).isdigit())]

# Convert to float
cl_df['Years of Experience'] = cl_df['Years of Experience'].astype(float)
cl_df.head()
```

```python
cl_df['NPV Value'] = cl_df['NPV Value'].str.extract(r'(\d+)').astype(float)
```

```python
cl_df['NPV Value'] = cl_df['NPV Value'].astype(float)
cl_df.head()
```

Dropping Duplicates

```python
cl_df.duplicated().sum()
```

```python
print(cl_df.count())
```

```python
cl_df = cl_df.drop_duplicates()
```

```python
print(cl_df.count())
```

Generating Doctor ID

```python
import itertools

letters = [''.join(c) for c in itertools.product('ABCDEFGHIJKLMNOPQRSTUVWXYZ', repeat=3)]
numbers = [f"{i:03}" for i in range(1, 1000)]

doctor_ids = [letter + num for letter in letters for num in numbers]

cl_df['Doctor_ID'] = doctor_ids[:len(cl_df)]

print(cl_df.head())
```

Shifting column

```python
col = cl_df.pop('Doctor_ID')
cl_df.insert(0, 'Doctor_ID', col)
cl_df.head()
```

```python
cl_df.to_csv('cleaned_data.csv', index=False)
```
